```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img/test2.jpg')
```

```python
plt.imshow(img)
```

```python
img_grayscaled = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_grayscaled, cmap='gray')
```

```python
img_blurred = cv2.GaussianBlur(img, (5,5), 3)
```

```python
plt.imshow(img_blurred, cmap='gray')
```

```python
kernelx = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
kernely = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
img_prewittx = cv2.filter2D(img_blurred, -1, kernelx)
img_prewitty = cv2.filter2D(img_blurred, -1, kernely)
```

```python
plt.imshow(img_prewittx)
```

```python
plt.imshow(img_prewitty)
```

```python
img_combined = cv2.addWeighted(src1=img_prewittx,alpha=0.5,src2=img_prewitty,beta=0.5,gamma=0)
```

```python
plt.imshow(img_combined)
```

```python
img_dilated = cv2.dilate(img_combined,np.ones((3,3)), iterations = 1)
```

```python
plt.imshow(img_dilated)
```

```python
closing = cv2.morphologyEx(img_combined, cv2.MORPH_CLOSE, np.ones((5,5)))
plt.imshow(closing)
```
